### Detecting and removing outliers

```python
outliers = []
def detect_outliers(data,threshold = 6):
    data = np.array(data, dtype = float)
    #threshold = 5
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i - mean) / std
        if abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pt
```

```python
def detect_outliers_in_columns(df, column_names, threshold=6):
    outliers_dict = {}
    for column_name in column_names:
        data = df[column_name]
        outliers = detect_outliers(data, threshold)
        outliers_dict[column_name] = outliers
    return outliers_dict

# Function to print outliers for each column
def print_outliers(outliers_dict):
    for column_name, outliers in outliers_dict.items():
        print(f"Outliers in column {column_name}: {outliers}")
```

```python
column_names = x.columns
```

```python
outliers_dict = detect_outliers_in_columns(x, column_names)
```

```python
print_outliers(outliers_dict)
```

```python
def remove_outliers(df, column_names, threshold=6):
    df_no_outliers = x.copy()
    for column_name in column_names:
        data = df[column_name]
        outliers = detect_outliers(data, threshold)
        
        # Create a boolean mask to filter out outliers
        mask = ~df[column_name].isin(outliers)
        df_no_outliers = df_no_outliers[mask]
    
    return df_no_outliers
```

```python
# Defining the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# initializing weights and bias
np.random.seed(0) # With the seed reset (every time), the same set of numbers will appear every time when we do np.random.rand(4).
w = np.random.randn(X_train.shape[1])
b = 0
```

```python
# Forward propagation : Compute predicted probabilites using the sigmoid fucntion
def predict(X,w,b):
    z = np.dot(X,w) + b
    return sigmoid(z)
```

```python
# Binary cross-entropy loss function
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-10 # to prevent log(0) errors
    return -np.mean(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))
```

```python
# Training Function
def Train(X_train,X_test,y_train,y_test,w,b,learning_rate = 0.01, epochs = 1000):
    train_losses = []
    test_losses = []

    for epoch in range(epochs):
        # Forward pass
        y_pred_train = predict(X_train, w, b)
        y_pred_test = predict(X_test, w , b)

        # Compute loss
        train_loss = binary_cross_entropy(y_train, y_pred_train)
        test_loss = binary_cross_entropy(y_test, y_pred_test)

        # Backward pass (Gradient descent)
        dw = np.dot(X_train.T , (y_pred_train - y_train)) / len(y_train)
        db = np.mean(y_pred_train - y_train)

        # update weights and bias
        w -= learning_rate * dw
        b -= learning_rate * db

        # Record losses
        train_losses.append(train_loss)
        test_losses.append(test_loss)

        # print progress
        if epoch % 100 == 0:
            print(f"Epoch {epoch}: Train Loss = {train_loss}, Test Loss = {test_loss}")

    return w , b , train_losses , test_losses
```

```python
# Training the model
w , b , train_losses, test_losses = Train(X_train, X_test, y_train, y_test, w ,b ,learning_rate = 0.01, epochs = 1000)
```

```python
# Plot training and testing losses
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('Training and Testing Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Binary Cross-Entropy Loss')
plt.legend()
plt.show()
```

```python
def make_predictions(X, w, b, threshold=0.5):
    probabilities = predict(X, w, b)
    return [1 if p >= threshold else 0 for p in probabilities]
```

```python
y_test_pred = make_predictions(X_test, w, b)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# Calculate and print evaluation metrics
accuracy = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred)
recall = recall_score(y_test, y_test_pred)
f1 = f1_score(y_test, y_test_pred)
conf_matrix = confusion_matrix(y_test, y_test_pred)

print(f"Test Set Accuracy: {accuracy * 100:.2f}%")
print(f"Test Set Precision: {precision:.2f}")
print(f"Test Set Recall: {recall:.2f}")
print(f"Test Set F1 Score: {f1:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```
